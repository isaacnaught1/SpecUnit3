#### Fill in blanks with a brief description of each plot type:

  
- Histogram
> a graph used to represent the frequency distribution of a a few data points of one variable
- Density Plot / KDE
> a representation of the distribution of a numeric variable
- Bar Chart
> presents categorical data with rectangular bars with heights or lengths proportional to the values they represent
- Line Plot
> displays information as a series of data points connected by straight line segments
- Univariate scatter plot
> displays the distribution of a series of quantitative values for different categories. shows all values across the series
- Violin Plot
> a hybrid of a box plot and a kernal density plot, shows peaks in data. depicts summary statistics and the density of each variable
- Box Plot
> shows range summary of a set of data by giving the minimum, first quartile, median, third quartile, and maximum values.
- Pie Chart
> a circle that has been deivided into sections that each represent a proportion of the whole
- Rug Plot
> plot of data for a single quantitative variable displayed as marks along an axis

### Step 5 - Create Plots for Each Variable in Dataset

```python
# import seaborn
import seaborn as sns

# set standard plot type as seaborn plot
sns.set()

# import matplotlib
import matplotlib.pyplot as plt

# write command for plots to appear in jupyter cells automatically
%matplotlib inline
```

```python
################################### TO DO #########################################
####### FOLLOW PROJECT INSTRUCTIONS TO COMPLETE THIS STEP IN THE CELLS BELOW ######
```

### Plot for 'age' column

```python
# create plot for age column
plt.figure(figsize=(12,6))

sns.violinplot(x=df['age'])

plt.xlabel("Age", fontsize=14)
plt.title("Density of Manager Ages")
```

##### Insights:

low to high 50's seem to have the highest density

### Plot for 'industry' column

```python
# create plot for industry column
plt.figure(figsize=(12,6))

ax = sns.countplot(data=df, x='industry',palette="icefire")

plt.xticks(rotation=90)
plt.xlabel("Industry", fontsize=13)
plt.ylabel("Number of Managers in Industry", fontsize=13)
plt.title("Industries in Survey", fontsize=15);
```

##### Insights:

About the same numbers across the industry

### Plot for 'salary' column

```python
# create plot for salary column
sns.displot(data = df, x = 'salary', kind = 'hist', bins=30)

plt.xlabel("Salary", fontsize=14)
plt.ylabel("Count", fontsize=14)
plt.title("Distribution of Salary Among Managers", fontsize=14)
```

##### Insights:

majorty make 80k to 120k

### Plot for 'industry_experience' column

```python
# create plot for industry experience column

plt.figure(figsize=(12,6))

ax = sns.histplot(data=df, x='industry_experience', bins=20)

plt.xlabel("Industry Experience ", fontsize=13)
plt.ylabel("Number of Managers", fontsize=13)
plt.title("Industry Experience Among Managers", fontsize=16)
```

##### Insights:

majority falls within 10-25 years

### Plot for 'education' column

```python
# create plot for education levels

plt.figure(figsize=(10,10))

data = df['education'].value_counts()

labels = ["Bachelor's Degree", "Master's Degree", "Some College", "High School"]

colors = sns.color_palette('Accent')

plt.title("Manager's Education Levels", fontsize=16)

ax = plt.pie(data, colors = colors, labels = labels, autopct = '%.0f%%')
```

##### Insights:

very few did not have some form of college education

### Plot for 'gender' column

```python
# create plot for gender column

plt.figure(figsize=(10,10))

data = df['gender'].value_counts()

labels = ['Male', 'Female']

colors = sns.color_palette('rainbow')

ax = plt.pie(data, colors=colors, labels=labels, autopct='%.0f%%')
```

##### Insights:

there is no majority between male and female managers

## Part 2 - Summary Statistics

### Step 1 - Review Summary Statistics

```python
################################### TO DO #########################################
####### FOLLOW PROJECT INSTRUCTIONS TO COMPLETE THIS STEP IN THE CELLS BELOW ######
```

- mean
> a single number representing the average of given data
- median
> the value in the middle 
- mode
> the value that is most commonly observed
- central tendency
> a value representing the center of its distribution to describe the data set
- distribution
> describes how values are common and uncommon
- variance
> a measurement of the spread between numbers in a data set
- standard deviation
> a measurement of how dispersed the data is in relation to the mean

#### Age Summary Statistics

```python
# use .describe to get summary statistics for age column
df['age'].describe()
```

```python
# create boxplot to visualize age summary statistics
plt.figure(figsize=(12,4))

plt.title("Age Distribution Among Managers", fontsize=15)

sns.boxplot(x=df['age'])

plt.xlabel("Manager Age", fontsize=13)
```

- age **median**

- age **minimum**

> The standard deviation is 12.75

### Step 3 - Find Salary Summary Statistics By Industry

```python
################################### TO DO #########################################
####### FOLLOW PROJECT INSTRUCTIONS TO COMPLETE THIS STEP IN THE CELLS BELOW ######
```

```python
# use groupby() and describe() to get salary summary statistics by industry
df['salary'].groupby(df['industry']).describe()
```

```python
# create visualization with multiple boxplots to show salary summary statistics by industry
plt.figure(figsize=(12,8))

ax = sns.boxplot(data=df, x='industry', y='salary')

plt.xlabel("Industry", fontsize=14)
plt.ylabel("Salary", fontsize=14)
plt.title("Distribution of Salaries Across Industries", fontsize=16)
```

##### Insights:

tech managers averages the most whilst health care managers averaged the least

### Step 4 - Find Salary Summary Statistics By Education

```python
# use groupby() and describe() to get salary summary statistics by education level
df['salary'].groupby(df['education']).describe()
```

```python
# create visualization with multiple boxplots to show salary summary statistics by education level
plt.figure(figsize=(12,8))

ax = sns.boxplot(data=df, x='education', y='salary')

plt.xlabel("Education", fontsize=14)
plt.ylabel("Salary", fontsize=14)
plt.title("Salary Distribution By Education Level", fontsize=16)
```

##### Insights:

The higher a person's education the higher their average salary

- **Categorical Data**

*Contingency Table*

>shows the frequency distribution of the values of the dependent variable

*Clustered Bar Chart*

>extends the bar chart, plotting numeric values for levels of two categorical variables instead of one.

- **Mixed Data**
*Individual Value Plot*

>shows a dot for the actual value of each observation in a group

*Strip Plot*
>a single-axis scatter plot that is used to visualise the distribution of many individual one-dimensional values

- **Quantitative Data**
*Scatter plots*

>shows the relationship between two quantitative variables measured for the same individuals
